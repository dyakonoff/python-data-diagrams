# Basic matplotlib diagrams - receipes

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
pd.options.display.max_rows    = 10
pd.options.display.max_columns = 20
pd.options.display.width       = 80

from IPython.display import display

# Check version number
print("pd version:", pd.__version__)
```

## Simple 2D functions

```python
def simple_graphs_sample():
    x = pd.Series( np.linspace(-1*np.pi, +1*np.pi, 100) )
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.tanh(x)
    plt.plot(x, y1, 'r--')
    plt.plot(x, y2, 'b');  # ';' suppress plot show object printed out in case of direct call in interactive mode
    plt.plot(x, y3, 'g-.')
    plt.show() # you can omit that in jupyter notebook & ipython
    
simple_graphs_sample()
```

## Scatter plot

```python
def generate_points(n):
    x1 = np.random.rand(n)*2 - 1
    x2 = np.random.rand(n)*2 - 1
    x = np.array([x1, x2]).T
    y = np.sign( x[:,1] - x[:,0] + 0.25*np.sin(np.pi*x[:,0]))
    return (x, y)

def test_generator():
    train_set = generate_points(500)
    x = train_set[0]
    c = train_set[1]
    cm = plt.cm.get_cmap('bwr')
    plt.scatter(x = x[:, 0], y = x[:, 1], c=c, cmap = cm, alpha = 0.7, label = []) 
    plt.legend()
    plt.grid(True)

    plt.show()
    
test_generator()
```
